## Question -1 :

__*Using the given  MT cars dataset you need to find outlier with z score__

```python
import pandas as pd
import seaborn as sns
import numpy as np
import os

# Change the directory where the dataset is available
os.chdir("D:\Dataset")

# Now import the dataset
data = pd.read_csv("mtcars.csv")

data
```

```python
data.shape
```

```python
sns.boxplot(x = data["hp"], data = data)
```

```python
Mean = np.mean(data["hp"])

SD = np.std(data["hp"])
```

```python
print("The mean of the data is :-",Mean)
print("The standard deviation is :-",SD)
```

```python
# Let's define a function that will find the outlier 

def outlier_checker(data):
    threshold = 2
    outlier = []
    for i in data:
        z_score = (i - Mean)/SD
        if z_score > threshold:
            outlier.append(i)
    print(outlier)
```

```python
# Now let's call the function that we just created

outlier_checker(data["hp"])

# AS we can see below that one outlier has been found
```

## Question - 2:

__* Using the Boston data set from sklearn, find out the Outlier and treat them__

```python
from sklearn.datasets import load_boston

boston = load_boston()
x = boston.data
y = boston.target
columns = boston.feature_names
```

```python
#create the dataframe
boston_data = pd.DataFrame(x,columns=columns)

boston_data
```

```python
boston_data.shape
```

```python
boston_data.describe()
```

```python
sns.boxplot(boston_data["DIS"])
```

```python
# This time we will follow the IQR approach to find the outliers

Q1 = boston_data["DIS"].quantile(0.25)
Q3 = boston_data["DIS"].quantile(0.75)

IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)
```

```python
# Set the upper and lower whiskers

Upper_whisker = Q3 + (1.5*IQR)
Lower_whisker = Q1 - (1.5*IQR)


print(Upper_whisker)
print(Lower_whisker)
```

```python
# lets loop through the data to find the outliers

for i in boston_data["DIS"]:
    if i > Upper_whisker:
        print(i)
    elif i < Lower_whisker:
        print(i)
        
# As we can see we have five outliers 
```

```python
# Now lets filter out these outliers from our dataset

new_data = boston_data[boston_data["DIS"] < Upper_whisker]

new_data.shape
```
